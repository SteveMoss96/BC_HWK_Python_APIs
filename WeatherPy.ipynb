```python
# create lists for holding lats and longs and cities
lat_long=[]
cities_list=[]
# Saying they should start at zero
i=0

# Randomly select at least 500 unique (non-repeat) cities based on latitude and longitude.
# Use np.random.uniform   https://numpy.org/doc/stable/reference/random/generated/numpy.random.uniform.html
# Use the zip() function returns a single tuple as per 12-Stu_UdemyZip and AskBCS: ask-97922
lats=np.random.uniform(low=-90.00, high=90.00, size =2000)
longs=np.random.uniform(low=-180.00, high=180.00, size =2000)
lat_long=zip(lats, longs)

# Use for loop to combine randomly generated lats and longs
for lat_long in lat_long:
    city = citipy.nearest_city(lat_long[0], lat_long[1]).city_name
    
    if city not in cities_list:
        cities_list.append(city)
    i += 1
    
# count the number of cities to see if have over 500
len(cities_list)    
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).

```python
# Set base url and units variables, and combine them to get the query url
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
query_url = f"{url}appid={weather_api_key}&units={units}&q="

# # # set up lists to hold reponse data for each field
city_id_list = []
city_name_list = []
country_list = []
lng_list = []
lat_list = []
temp_list = []
humidity_list = []
clouds_list = []
wind_speed_list = []

# # Loop through the list of cities and perform a request for data on each

for city in cities_list:
    response = requests.get(query_url + city).json()
#     print(response)
    try:
        city_id_list.append(response["id"])
        city_name_list.append(response['name'])
        country_list.append(response['sys']['country'])
        lng_list.append(response['coord']['lon'])
        lat_list.append(response['coord']['lat'])
        temp_list.append(response['main']['temp'])
        humidity_list.append(response['main']['humidity'])
        clouds_list.append(response['clouds']['all'])
        wind_speed_list.append(response['wind']['speed'])
        print(f"City Name: {city}, City ID: {response['id']}")
# in exception below 'e' just stands for the exception you are going to except. Common code phrase usage
    except Exception as e:
        print(e)
   
```

```python
# separate them into Northern Hemisphere (>=  0 degrees latitude) and Southern Hemisphere (< 0 degrees latitude)
# use the tilde operator - ~  NOT Inverts all the bits
# Create Northern and Southern Hemisphere DataFrames
N_or_S = CityWeath_df['Lat'] >= 0
north_hemis = CityWeath_df[N_or_S]
south_hemis = CityWeath_df[~N_or_S]
# north_hemis.head()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = north_hemis["Lat"]
y_values = north_hemis["Temperature"]

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,4)) + "x + " + str(round(intercept,4))
plt.scatter(x_values,y_values, marker = 'D', color="violet", edgecolors="black")
plt.plot(x_values,regress_values,"--", color="black")
plt.title("Northern Hemisphere - Max Temp vs. Latitude Linear Regression")
plt.xlabel("Latitude")
plt.ylabel("max temperature")
# plt.ylim([33, 47])
# plt.xlim([14.5, 25.25])
# # Annotate linear regression
plt.annotate(line_eq,(10,15),fontsize=10,color="purple")
plt.show()
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = south_hemis["Lat"]
y_values = south_hemis["Temperature"]

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,4)) + "x + " + str(round(intercept,4))
plt.scatter(x_values,y_values, marker = 'D', color="yellow", edgecolors="black")
plt.plot(x_values,regress_values,"--", color="black")
plt.title("Southern Hemisphere - Max Temp vs. Latitude Linear Regression")
plt.xlabel("Latitude")
plt.ylabel("max temperature")
# plt.ylim([33, 47])
# plt.xlim([14.5, 25.25])
# # Annotate linear regression
plt.annotate(line_eq,(-50,94),fontsize=10,color="purple")
plt.show()
```
